```python
import numpy as np
import matplotlib.pyplot as plt
from hmmlearn import hmm
import pandas as pd
import zipfile
from sklearn.metrics import accuracy_score

with zipfile.ZipFile(r"C:\Users\HP\Downloads\archive.zip", 'r') as z:
    credit_card = pd.read_csv(z.open('creditcard_2023.csv'))

X = credit_card.drop(['id', 'Class'], axis=1).values  # Convert to numpy array for HMM
y = credit_card['Class'].values  # Labels

model = hmm.GaussianHMM(n_components=2, covariance_type="diag", n_iter=100)
model.fit(X)

hidden_states = model.predict(X)

state_means = np.array([model.means_[i] for i in range(model.n_components)])
print("State Means:\n", state_means)

state_to_class = {0: 0, 1: 1}  
predictions = np.array([state_to_class[state] for state in hidden_states])
```

```python
# Calculate accuracy
accuracy = accuracy_score(y, predictions)
print(f'Accuracy of the HMM model: {accuracy:.2f}')
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.title('Observed Data - Feature: Amount')
plt.plot(X[:, -2], label='Amount', color='blue')  # Assuming 'Amount' is the second last column
plt.xlabel('Time Steps')
plt.ylabel('Amount Value')
plt.legend()

# Plot the hidden states
plt.subplot(2, 1, 2)
plt.title('Hidden States')
plt.plot(hidden_states, label='Hidden States', color='orange')
plt.xlabel('Time Steps')
plt.ylabel('State')
plt.yticks(np.arange(2))  # Two hidden states
plt.legend()

plt.tight_layout()
plt.show()
```
